```python
def balanced(number):

    iter = 0
    mid_count = int(len(str(number))) / 2
    left_sum = 0
    right_sum = 0
    lenght = int(len(str(number)))

    left_side = ''
    right_side = ''
    counter = 0

    result = ''
    palindrome_count = 0


    if len(str(number)) % 2 != 0:
        mid_count -= 1
        

    for x in str(number):
        if iter < mid_count:
            left_sum += int(x)
            left_side += x
            iter += 1 
            

    if len(str(number)) % 2 != 0:
        iter += 1

    split_range = str(number)[iter:lenght]

    for y in split_range:
        right_sum += int(y)
        right_side = y + right_side

    if right_side == left_side:
        return number
     
    else:
        return None



    

def count_the_polindrome(num):
    list = []
    sum_polindrome_count = 0
    counter = 10
   
    while counter <= int(num):
        list += [counter]
        counter += 1

    for x in list:
        element = balanced(x)

        if element != None:
            sum_polindrome_count += 1
        else:
            sum_polindrome_count += 0
            

       
    return sum_polindrome_count

  
        
        

        


            

tests = [
    (10, 0),
    (20, 1),
    (30, 2),
    (101, 10),
    (200, 19),
    (43539, 525),
    (4247, 132),
    (48877, 577),
    (94012, 1029),
    (62560, 715),
    (92009, 1009),
    (63176, 721),
    (67409, 763),
    (62834, 718),
    (77420, 863),
    (99999, 1089),
]


for n, expected in tests:
    result = count_the_polindrome(n)

    print(result == expected)



  
```
